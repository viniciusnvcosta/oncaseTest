### Insight 1:
O valor `target` de fato não apresenta uma distribuição normal, podemos testar o _fit_ em algumas distribuições conhecidas e indentificar o seu comportamento

```python
f = Fitter(
    df_viz["target"].values, distributions=["gamma", "lognorm", "beta", "burr", "norm"]
)
f.fit()
f.summary()
```

```python
for i in df_viz[["X7", "X2", "X1"]].columns:
    qq_sample = df_viz[i].sample(frac=0.2, random_state=42)

    z_score = (qq_sample - np.mean(qq_sample)) / np.std(qq_sample)

    ss.probplot(z_score, dist="norm", plot=plt)
    plt.title(i + " Q-Q plot")
    plt.figure(figsize=(2, 2))
    plt.show()
```

> As variáveis `X2` e `X7` selecionadas como features têm comportamento de Gaussiana e então podem ser padronizadas com o `Standard Scaler`

## Data Preprocessing

##### split dos valores alvo de treino e teste

```python
X_train_raw = df_train.loc[:, ["X2", "X7"]]
X_train_raw.to_numpy()
display(X_train_raw.head(1))
y_train_raw = df_train["target"].values.reshape(-1, 1)

X_test_raw = df_test.loc[:, ["X2", "X7"]]
X_train_raw.to_numpy()
display(X_test_raw.head(1))
# y_test_raw => y_test_missing 
y_test_missing = df_test["target"].values.reshape(-1, 1)
```

    como não vamos remover os outliers da base de teste, vamos seguir a taxonomia e declarar 'y_test_raw' como 'y_test_missing'

#### remover outliers

```python
def outlier_iqr(dados, feature):
    Q1 = dados[feature].quantile(0.25)
    Q3 = dados[feature].quantile(0.75)
    IQR = Q3 - Q1
    LI = Q1 - 1.5 * IQR
    LS = Q3 + 1.5 * IQR
    outliers = dados[(dados[feature] < LI) | (dados[feature] > LS)]
    df_viz.boxplot(column=feature, return_type="axes", figsize=(6, 4), grid=True)
    plt.show()
    return outliers
```

```python
outliers = set()
for i in X_train_raw.columns:
    v = outlier_iqr(X_train_raw, i)
    outliers.update(v.index)
    print("nº de outliers", len(v), "\n========")
```

```python
print(
    "o percentual de outliers em treino é: ",
    round((len(outliers) / len(X_train_raw)) * 100, 2),
    "%",
)
```

removendo os outliers

```python
X_train_raw.drop(index=outliers, inplace=True)
```

```python
y_train_missing = np.delete(y_train_raw, list(outliers), axis=0)
```

### Pipeline de Pré-processamento

```python
len(y_train_missing)
```

```python
len(X_train_raw)
```

```python
KNN_imp = KNNImputer(missing_values=np.nan, n_neighbors=5)
```

```python
y_train = KNN_imp.fit_transform(y_train_missing)
y_test = KNN_imp.fit_transform(y_test_missing)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.fit_transform(X_test_raw)
```

### Regressão Linear Múltipla

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

y_pred_train = linear_regressor.predict(X_train)
y_pred_test = linear_regressor.predict(X_test)
```

#### métricas

```python
mae = mean_absolute_error(y_test, y_pred_test)
mape = mean_absolute_percentage_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
rmse = mean_squared_error(y_test, y_pred_test, squared=False)
cv_score = cross_val_score(estimator=linear_regressor, X=X_train, y=y_train, cv=10)
r2_score_train = r2_score(y_train, y_pred_train)
r2_score_test = r2_score(y_test, y_pred_test)
```

```python
print("MAE: ", mae)
print("MAPE: ", mape)
print("MSE: ", mse)
print("RMSE: ", rmse)
print("R2_score (train): ", r2_score_train)
print("CV *R2_score (train): ", cv_score.mean())
print("R2_score (test): ", r2_score_test)

plt.figure(figsize=(12, 8))
plt.plot(X_train, y_train, "bo")
plt.plot(X_test, y_pred_test, "r.")
plt.show()
```

```python
linear_regressor.score(X_test, y_test)
```
